```python
import pandas as pd

# Load the datasets
iconan_train = pd.read_csv('data/iconan/iconan_train.csv')
iconan_test = pd.read_csv('data/iconan/iconan_test.csv')
iconan_dev = pd.read_csv('data/iconan/iconan_dev.csv')
iconanv2_train = pd.read_csv('data/iconanv2/train.csv')
iconanv2_test = pd.read_csv('data/iconanv2/test.csv')
iconanv2_val = pd.read_csv('data/iconanv2/val.csv')

# Combine data for analysis
iconan = pd.concat([iconan_train, iconan_test, iconan_dev])
iconanv2 = pd.concat([iconanv2_train, iconanv2_test, iconanv2_val])

# Group by target and count the rows for each target in ICONAN
iconan_target_distribution = iconan['Target'].value_counts()

# Group by target and count the rows for each target in IntentConanV2
iconanv2_target_distribution = iconanv2['hatespeechTarget'].value_counts()

# Print the results
print("Target Distribution in ICONAN:")
print(iconan_target_distribution)
print("\nTarget Distribution in IntentConanV2:")
print(iconanv2_target_distribution)
```

```python
import matplotlib.pyplot as plt

# Plot Target Distribution in ICONAN
plt.figure(figsize=(14, 6))
iconan_target_distribution.plot(kind='bar', color='skyblue')
plt.title('Target Distribution in ICONAN')
plt.xlabel('Hate Speech Target')
plt.ylabel('Number of Instances')
plt.xticks(rotation=45)
plt.show()

# Plot Target Distribution in IntentConanV2
plt.figure(figsize=(14, 6))
iconanv2_target_distribution.plot(kind='bar', color='salmon')
plt.title('Target Distribution in IntentConanV2')
plt.xlabel('Hate Speech Target')
plt.ylabel('Number of Instances')
plt.xticks(rotation=45)
plt.show()
```
